```python
import os
import pandas as pd
```

```python
dfs = []
for f in os.listdir():
    if f.endswith('xlsx'): 
        dfs += [pd.read_excel(f,index_col=None)]
len(dfs)
```

```python
print(dfs[0].shape)
dfs[0].head()
```

```python
print(dfs[1].shape)
dfs[1].head()
```

```python
print(dfs[2].shape)
dfs[2].head()
```

```python
print(dfs[3].shape)
dfs[3].head()
```

```python
dfs[0] = dfs[0].rename(columns={'Stkcd':'Symbol','Reptdt':'EndDate'})
dfs[0].head()
```

```python
dfs[1] = dfs[1].rename(columns={'Stkcd':'Symbol','Reptdt':'EndDate'})
dfs[1].head()
```

```python
max_row, total_column=max([v.shape[0] for v in dfs]),sum([v.shape[1] for v in dfs])
max_row,total_column
```

```python
df = dfs[0].merge(dfs[1],how='outer').merge(dfs[2],how='outer').merge(dfs[3],how='outer')
print(df.shape)
df.head()
```

```python
df.to_excel('out_join.xlsx')
```
